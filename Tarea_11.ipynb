```julia
#Creamos la función fuyos argumentos son la función f, el valor inicial x0, los extremos del intervalo en t (t0 y tf) y el tamaño del intervalo (h)
function metodo_euler(f,x0,t0,tf,h)
#n es el número de elementos de la partición que debemos hacer del intervalo (t0,tf)
   n=round((tf-t0)/h)+1
    
    #usamos linspace para dividr (t0,tf) en n elementos y guardamos esto como listt
    listt=linspace(t0,tf,n)
    
    #creamos un arreglo de n zeros, al que llamamos listx
    listx=zeros(n)
 
    #definimos el primer elemento como nuestro valor inicial
 listx[1]=x0
    
#el siguiente for es para aplicar la relación de recurrencia
    for i in 1:length(listx)-1
        listx[i+1]=listx[i]+h*f(listx[i],listt[i])
    end
#la función regresa un arreglo, cuyos elementos son los arreglos listt y listx
    return listt, listx
end;
```

Queremos resolver $$\frac{dx}{dt}=-\alpha x,$$ con $\alpha= -2.5$ y $t\in(0,8)$.

```julia
#Definimos la función que representa la derivada
f(x,t)= -2.5x;
```

a) Primero resolvamos con la condición inicial $x(0)=1$.

```julia
#Para h=1.0
metodo_euler(f,1,0,8,1.0)
```

```julia
#Para h=0.5
metodo_euler(f,1,0,8,0.5)
```

```julia
#Para h=0.2
metodo_euler(f,1,0,8,0.2)
```

```julia
#Para h=0.01
metodo_euler(f,1,0,8,0.01)
```

Con este condición inicial, se obtiene que la solución exacta es:
$$x(t)=e^{-2.5 t}.$$ La cual usaremos para comparar la convergencia de las diferentes $h's$.

```julia
using PyPlot
```

```julia
#Usamos plot para graficar las soluciones con las diferentes h's (a excepción de h=1) y compararlas con la solución analítica.

#Recordando que metodo_euler nos regresa un arreglo con dos entradas, para el eje de las abscisas tomamos la primera entrada (listt) y para el eje de las ordenadas tomamos la segunda entrada (listx).
plot(metodo_euler(f,1,0,8,0.5)[1],metodo_euler(f,1,0,8,0.5)[2],label="h=0.5");
plot(metodo_euler(f,1,0,8,0.2)[1],metodo_euler(f,1,0,8,0.2)[2],label="h=0.2");
plot(metodo_euler(f,1,0,8,0.01)[1],metodo_euler(f,1,0,8,0.01)[2],label="h=0.01");

#Para graficar la solución analítica, dividimos el intervalo (0,8) en 100 puntos y aplicamos la solución (mostrada arriba). Usamos .^ para indicar que estamos operando sobre arreglos
plot(linspace(0,8,100),e.^(-2.5(linspace(0,8,100))),label="Solución analítica");
xlabel("t");
ylabel("x(t)");
legend();
show();
```

```julia
#Graficamos de forma independiente para h=1, pero llevando a cabo los mismos pasos.

plot(metodo_euler(f,1,0,8,1)[1],metodo_euler(f,1,0,8,1)[2],label="h=1");
plot(linspace(0,8,100),e.^(-2.5(linspace(0,8,100))),label="Solución analítica");
xlabel("t");
ylabel("x(t)");
legend();
show();
```

__Ejercicio 2:__

Consideremos una partícula puntual de masa $m$ que se encuentra sobre el eje x. Adicionalmente, supongamos que dicha partícula se encuentra sujeta a un resorte cuya constante es $k'$ y cuyo punto de equilibro ocurren en $x=0$. Por la segunda ley de Newton y por ley de Hooke, tenemos que la fuerza ejercida sobre la partícula es
$$F=m\ddot{x}=-k'x.$$
Si definimos $k=\frac{k'}{m}$, entonces
$$\ddot{x}=-kx.$$

Ahora, procedemos a resolver la ecuación. Proponemos como solución $x(t)=e^{mt}$, con $m$ por determinar. De este modo, $\ddot{x}=m^{2}e^{mt}$. Sustituyendo en la ecuación original, se tiene que 

$$m^{2}e^{mt}+ke^{mt}=0.$$

Como $e^{mt}\neq 0$ para cualquier $t$ y $m$, entonces podemos dividir entre eso. Así,

$$m^{2}+k=0$$

$$\implies m=\pm i\sqrt{k}.$$

De lo anterior, tenemos las soluciones linealmente independientes:

$$x_{1}(t)=e^{i\sqrt{k}t},$$
$$x_{2}(t)=e^{-i\sqrt{k}t}.$$

Entonces, la solución general será

$$x(t)=C e^{i\sqrt{k}t}+D e^{-i\sqrt{k}t},$$
con $C$ y $D$ constantes.

Podemos reescribir la expresión anterior en términos de senos y cosenos de la siguiente forma:

$$x(t)=Asin(\sqrt{k}t)+ B cos(\sqrt{k}t).$$
Donde $A$ y $B$ son constantes reales (pues buscamos soluciones reales).

Por un lado, veamos que
$$x(0)=B.$$

Por otro lado, $\dot{x}(t)=A\sqrt{k}cos({k}t)-B\sqrt{k}sin(\sqrt{k}t)$, por lo que

$$\dot{x}(0)=A\sqrt{k}.$$

Por lo tanto, $$x(t)=x(0)cos(\sqrt{k}t) + \frac{\dot{x}(0)}{\sqrt{k}}sin(\sqrt{k}t).$$

__Ejercicio 3:__

Definimos $u(t)=\dot{x}(t)$.

De este modo, la derivada de $u$ será $\dot{u}(t)=\ddot{x}(t)$

Entonces, sustituyendo en la ecuación original, $$\dot{u}(t)=\ddot{x}(t)=-kx(t)$$

$$\implies \dot{u}(t)=-kx(t).$$

Por lo tanto, tenemos el siguiente sistema de ecuaciones:

$$\dot{x}=u$$
$$\dot{u}=-kx.$$

__Ejercicio 4:__

```julia
#Creamos la función que aplica el método de Euler para cualquier dimensión. Los argumentos son la función f, el arreglo de valores inciales x0, los extremos del intervalo en t (t0 y tf) y el tamaño del intervalo (h)

function metodo_euler_dim(f,x0,t0,tf,h)
#n es el número de elementos de la partición que debemos hacer del intervalo (t0,tf)
   n=round((tf-t0)/h)+1
    
    #usamos linspace para dividr (t0,tf) en n elementos y guardamos esto como listt
    listt=linspace(t0,tf,n)
    
#Definimos x como nuestro valor inicial (arreglo de valores iniciales)
    x=x0
#Definimos listx como un objeto de tipo indefinido. Aquí es donde iremos guardando los puntos de la solución.
    listx=[]
    
#Guardamos nuestro valor inicial en listx
    push!(listx,x)

#el siguiente for es para aplicar la relación de recurrencia
    for i in 1:length(listt)-1
        
#Recordemos que el valor inicial de x es x0, por lo que el rango de i es correcto.
        x = x + f(x,listt[i])*h
        
#Cada que aplicamos la relación de recurrencia, vamos guardando el punto en listx.
        push!(listx,x) 
    end
#la función regresa un arreglo, cuyos elementos son los arreglos listt y listx
    return listt, listx
end;
```

Queremos resolver $\ddot{x}(t)=-kx(t).$ Como se vio en el ejercicio 3, esta ecuación es quivalente al sistema de ecuaciones acopladas:
$$\dot{x}=u$$
$$\dot{u}=-kx.$$

Usaremos el método de Euler en su forma vectorial, donde se tiene la expresión $$\frac{d\vec{x}}{dt}=\vec{g}(\vec{x},t).$$

En este caso $\vec{x}=(x,u)$ y $\vec{g}(\vec{x},t)=(u,-kx)$. 

En este caso, por simplicidad, escojemos $k=1$. Además, escogemos $\vec{x}(0)=(x(0),u(0))=(5,2)$ y $t \in (0,10).$

```julia
g(x,t)=[x[2],-x[1]];
```

Ahora probamos el método para diferentes $h's$.

```julia
#Para h=1
metodo_euler_dim(g,[5,2],0,10,1)
```

```julia
#Para h=0.5
metodo_euler_dim(g,[5,2],0,10,0.5)
```

```julia
#Para h=0.2
metodo_euler_dim(g,[5,2],0,10,0.2)
```

```julia
#Para h=0.01
metodo_euler_dim(g,[5,2],0,10,0.01)
```


Por otra parte, por el ejercicio 3, sabemos que la solución analítica es $$x(t)=5cos(t)+2sin(t).$$
La cual usaremos para comparar la convergencia con las diferentes $h's$.

__Nota__: Cuando evaluamos metodo_euler_dim, vemos que nos regresa un arreglo con dos elementos. El primer elemento es el arreglo de la partición del intervalo en $t$, mientras que el segundo elemento es un arreglo de parejas, donde la primer entrada es el valor de x y la segunda el valor de u. En este caso, sólo nos interesa graficar x.

```julia
#Usamos plot para graficar las soluciones con las diferentes h's (a excepción de h=1) y compararlas con la solución analítica.

#Justificandonos en la nota anterior el eje de las abscisas lo tomamos como la primera entrada del arreglo que nos regresa metodo_euler_dim. Mientras que en el eje de las ordenadas, tomamos en cuenta la segunda entrada de metodo_euler_dim, pero usamos map para obtener una lista únicamente de las primeras entradas de cada pareja.

plot(metodo_euler_dim(g,[5,2],0,10,0.5)[1],map(x->x[1],metodo_euler_dim(g,[5,2],0,10,0.5)[2]),label="h=0.5");
plot(metodo_euler_dim(g,[5,2],0,10,0.2)[1],map(x->x[1],metodo_euler_dim(g,[5,2],0,10,0.2)[2]),label="h=0.2");
plot(metodo_euler_dim(g,[5,2],0,10,0.01)[1],map(x->x[1],metodo_euler_dim(g,[5,2],0,10,0.01)[2]),label="h=0.01");

#Para graficar la solución analítica, dividimos el intervalo (0,10) en 100 puntos y aplicamos la solución (mostrada arriba). Usamos sin.() y cos.() para indicar que estamos operando sobre arreglos

plot(linspace(0,10,100),5*cos.(linspace(0,10,100))+2*sin.(linspace(0,10,100)),label="Solución analítica");

xlabel("t");
ylabel("x(t)");
legend();
show();
```

Notamos que para $h=0.01$ la gráfica de la solución que nos da el método coincide de forma muy precisa con la de la solución analítica.

```julia
#Graficamos de forma independiente para h=1, pero llevando a cabo los mismos pasos.
plot(metodo_euler_dim(g,[5,2],0,10,1)[1],map(x->x[1],metodo_euler_dim(g,[5,2],0,10,1)[2]),label="h=1");

plot(linspace(0,10,100),5*cos.(linspace(0,10,100))+2*sin.(linspace(0,10,100)),label="Solución analítica");

xlabel("t");
ylabel("x(t)");
legend();
show();
```

Queremos resolver la ecuación $$\frac{dx}{dt}=\frac{2x-18 t}{1+t}.$$

Cuya solución analítica, sabemos que es $x(t)=4+8t-5t^2$, cuando $x(0)=4$.
Además escogemos $t\in(0,12)$.

```julia
#definimos la función asociada a la derivada
w(x,t)=(2*x-18*t)/(1+t);
```

```julia
#Para h=1.0
metodo_euler(w,4,0,12,1.0)
```

```julia
#Usamos plot para graficar las soluciones con las diferentes h's (a excepción de h=1) y compararlas con la solución analítica.

#Recordando que metodo_euler nos regresa un arreglo con dos entradas, para el eje de las abscisas tomamos la primera entrada (listt) y para el eje de las ordenadas tomamos la segunda entrada (listx).
plot(metodo_euler(w,4,0,12,0.5)[1],metodo_euler(w,4,0,12,0.5)[2],label="h=0.5");
plot(metodo_euler(w,4,0,12,0.2)[1],metodo_euler(w,4,0,12,0.2)[2],label="h=0.2");
plot(metodo_euler(w,4,0,12,0.01)[1],metodo_euler(w,4,0,12,0.01)[2],label="h=0.01");

#Para graficar la solución analítica, dividimos el intervalo (0,12) en 100 puntos y aplicamos la solución (mostrada arriba). Usamos .^ para indicar que estamos operando sobre arreglos

list=linspace(0,12,100)
plot(list,4+8*list-5*list.^(2),label="Solución analítica");
xlabel("t");
ylabel("x(t)");
legend();
show();
```

Nuevamente observamos que la solución se acerca cada vez más mientras menor sea $h$. En particular, para $h=0.01$, las gráficas ya son muy similares.

```julia
#Graficamos de forma independiente para h=1, pero llevando a cabo los mismos pasos.

plot(metodo_euler(w,4,0,12,0.5)[1],metodo_euler(w,4,0,12,0.5)[2],label="h=1");
plot(list,4+8*list-5*list.^(2),label="Solución analítica");
xlabel("t");
ylabel("x(t)");
legend();
show();
```
