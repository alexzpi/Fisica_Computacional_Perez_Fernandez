__Ejercicio 2__

Tenemos la ecuación:
$$\frac{d^{2}f}{dx^{2}}=\rho (x)=\cos(x)$$

```julia
#Hacemos 12 particiones en el intervalo de 0 a 2pi
xlist=linspace(0,2pi,12);
#La distancia h entre cada punto es
h=xlist[2]-xlist[1];
```

Usaremos las condiciones en la frontera:
$$f(0)=f(2\pi)=1$$

```julia
#Damos una "advinanza" para los valores de f (todos igual a 1)
f=zeros(12)
for i in 1:length(f)
    f[i]=1
end
```

Para el método de relajación de una dimensión se tiene que:
$$f_{i}=\frac{1}{2}\left(f_{i+1}+f_{i-1}-h^2\rho(x_{i})\right)$$

```julia
#Aplicamos el método de relajación, iterando 1000 veces

for k in 1:1000
    for i in 2:length(f)-1
    
        f[i]=(1/2)*(f[i+1]+f[i-1]-(h^2)cos(xlist[i]))
    end
end
```

Esta ecuación diferencial se puede resolver explícitamente, dando:
$$f(x)=c_{1}x+c_{2}-\cos(x)$$
Específicamente, para las condiciones de frontera propuestas, $c_{1}=0$ y $c_{2}=2$. Así:
$$f(x)=2-\cos(x)$$

```julia
#Graficamos la solución
plot(xlist,f,label="Solución con método de relajación");
#Graficamos la solución analítica
a=linspace(0,2pi,100)
plot(a,2-cos.(a), label="Solución analítica")
legend()
xlabel("x")
ylabel("f")
show();
```

__Ejercicio 3__

Primero recordemos los métodos de Euler y de Runge-Kutta de orden 4:

```julia
#Definimos el método de Euler para funciones de cualquier dimensión. Las entradas son: f, la función que nos da la derivada;listt,el arreglo de t's; y x0, el valor inicial.

function euler(f,listt,x0)
    
#Definimos h como la distancia entra las dos primeras entradas de listt.
    h=listt[2]-listt[1]
    
#Definimos x como nuestro valor inicial (arreglo de valores iniciales)
    x=x0
#Definimos listx como un objeto de tipo indefinido. Aquí es donde iremos guardando los puntos de la solución.
    listx=[]
    
#Guardamos nuestro valor inicial en listx
    push!(listx,x)

#el siguiente for es para aplicar la relación de recurrencia
    for i in 1:length(listt)-1
        
#Recordemos que el valor inicial de x es x0, por lo que el rango de i es correcto.
        x = x + f(x,listt[i])*h
        
#Cada que aplicamos la relación de recurrencia, vamos guardando el punto en listx.
        push!(listx,x) 
    end
#la función regresa el arreglo de x's
    return listx
end;






#Definimos el método de Runge-Kutta de orden 4 para funciones de cualquier dimensión. Las entradas son: f, la función que nos da la derivada;listt,el arreglo de t's; y x0, el valor inicial.
function rk_4(f,listt,x0)
    
#Definimos h como la distancia entra las dos primeras entradas de listt.
    h=listt[2]-listt[1]
#Definimos x como nuestro valor inicial (arreglo de valores iniciales)    
    x=x0
    
#Definimos listx como un objeto de tipo indefinido. Aquí es donde iremos guardando los puntos de la solución.
    listx=[]
    
#Guardamos nuestro valor inicial en listx
    push!(listx,x)

#el siguiente for es para aplicar la relación de recurrencia
    for i in 1:length(listt)-1
        k1=f(x,listt[i])
        k2=f(x+(h/2)*k1,listt[i]+h/2)
        k3=f(x+(h/2)*k2,listt[i]+h/2)
        k4=f(x+h*k3,listt[i+1])
        
        x=x+(h/6)*(k1+2*k2+2*k3+k4)
        
 #Cada que aplicamos la relación de recurrencia, vamos guardando el punto en listx.       
        push!(listx,x)
    end
#la función regresa el arreglo de x's
    return listx
end;
```

```julia
#Dividimos el intervalo de 0 a 1 en 12 puntos
x_list=linspace(0,1,12)
#Definimos h_x como la distancia entre cada uno de dichos puntos
h_x=x_list[2]-x_list[1];
```

Tenemos la ecuación de calor:
$$\frac{\partial^{2}f}{\partial x^{2}}-D\frac{\partial f}{\partial t}=0.$$

De forma discreta, podemos escribir:
$$\frac{f_{i+1}^{[k]}-2f_{i}^{[k]}+f_{i-1}^{[k]}}{h_{x}^{2}}=D\frac{f_{i}^{[k+1]}-f_{i}^{[k]}}{h_{t}},$$
donde el subínidice $i$ indica los puntos en la malla (en este caso una recta) y el superíndice $[k]$ indica la evolución temporal.

De este modo, si vemos a $f$ como un vector cuyas entradas son las $f_{i}'s$, entonces tenemos que

$$\frac{d f_{i}}{d t}= \frac{f_{i+1}^{[k]}-2f_{i}^{[k]}+f_{i-1}^{[k]}}{D h_{x}^{2}}$$

Por lo tanto podemos usar tanto el método de Euler como el de Runge-Kutta (para varias dimensiones), para resolver el problema.

En este caso usaremos $D=1.$

```julia
#Definimos la función que caracteriza la evolución temporal del sistema (mostrada anteriormente)

function F(f,t::Float64)
    F=zeros(12)
   for i in 2:11
        F[i]=(h_x^(-2))*(f[i+1]-2*f[i]+f[i-1])
    end
    return F
end; 
```

```julia
#Definimos las condiciones iniciales
f_0=[10,5,4,3,2,1,3,5,7,7,8,0];
timelist=0:0.001:10;
```

a) Resolvemos por método de Euler

```julia
#Aplicamos el método de Euler
sol_euler=euler(F,timelist,f_0);
```

```julia
#Graficamos para los primeros tiempos
for i in 1:10
    plot(x_list,sol_euler[i])
end

xlabel("x")
ylabel("f(x,t)")
show();
```

```julia
#Graficamos para tiempos grandes

for i in length(timelist)-1000:length(timelist)
    plot(x_list,sol_euler[i])
end

xlabel("x")
ylabel("f(x,t)")
show();
```

b) Resolvemos con Runge-Kutta

```julia
#aplicamos runge-kutta de orden 4
sol_rk=rk_4(F,timelist,f_0);
```

```julia
#Graficamos para los primeros tiempos
for i in 1:10
    plot(x_list,sol_rk[i])
end

xlabel("x")
ylabel("f(x,t)")
show();
```

```julia
#Graficamos para tiempos grandes

for i in length(timelist)-1000:length(timelist)
    plot(x_list,sol_rk[i])
end

xlabel("x")
ylabel("f(x,t)")
show();
```

Notamos que en ambos casos (Euler y Runge-Kutta) las soluciones para tiempos grandes se aproximan a una recta, lo cual era de esperarse, ya que cuando $t\rightarrow \infty$, las soluciones de la ecuación de calor tienden a soluciones estacionarias (soluciones de la ecuación de Laplace), que en una dimensión se trata de rectas.
